```python
def change_price(a,b):
  low = []
  med = []
  high = []
  for i,row in data.iterrows():
    if(int(row['Price'])<=a):
      low.append(i)
    if(int(row['Price'])> a and int(row['Price'])<= b):
      med.append(i)
    if int(row['Price']) > b :
      high.append(i)
  print(len(low))
  print(len(med))
  print(len(high))
  data.loc[low,'Price'] = data.loc[low,'Price'].map(lambda x : 'LOW')
  data.loc[med,'Price'] = data.loc[med,'Price'].map(lambda x : 'MEDIUM')
  data.loc[high,'Price'] = data.loc[high,'Price'].map(lambda x : 'HIGH')

def change_size(a,b):
  low = []
  med = []
  high = []
  for i,row in data.iterrows():
    if(int(row['Size'])<=a):
      low.append(i)
    if(int(row['Size'])> a and int(row['Size'])<= b):
      med.append(i)
    if int(row['Size']) > b :
      high.append(i)
  print(len(low))
  print(len(med))
  print(len(high))
  data.loc[low,'Size'] = data.loc[low,'Size'].map(lambda x : 'LOW')
  data.loc[med,'Size'] = data.loc[med,'Size'].map(lambda x : 'MEDIUM')
  data.loc[high,'Size'] = data.loc[high,'Size'].map(lambda x : 'HIGH')
```

```python
def logic_formula(R):
  f = ""
  for i,row in R.iterrows():
    tuple_f = ""
    for j,x in enumerate(row):
      tuple_f += f"({R.columns[j]} = {str(x)})"
      if j != len(row)-1:
        tuple_f += ' ^ '
    f += f"{tuple_f}"
    if i != len(R)-1 :
      f+= 'U'
  return f
```
