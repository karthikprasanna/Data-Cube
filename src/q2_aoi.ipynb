```python
# Import the pandas library
import pandas as pd

# Load the dataset from the Excel file
data = pd.read_excel("../data/q2data.xlsx")

# Display the first few rows of the dataset to inspect the data
print(data.head())
```

```python
data
```

```python
selected_columns = [
    "Property_type",
    "Property_status",
    "Price_per_unit_area",
    "Property_building_status",
    "City_name",
    "No_of_BHK",
    "Price",
    "Size",
    "is_furnished",
    "is_ready_to_move"
]

filtered_data = data[selected_columns]

# Display the first few rows of the filtered DataFrame
print(filtered_data.head())
```

```python
# Filter the DataFrame to include only rows where Property_building_status is 'ACTIVE'
filtered_data = filtered_data[filtered_data["Property_building_status"] == "ACTIVE"]

# Display the first few rows of the filtered DataFrame
# print(filtered_data.head())
filtered_data
```

```python
# Clean 'Size' column by removing non-numeric characters
filtered_data['Size'] = filtered_data['Size'].str.replace(',', '').str.replace(' sq ft', '').astype(float)

# Clean and convert other columns
filtered_data['Price'] = filtered_data['Price'].str.replace(',', '').astype(float)
# Convert 'Price_per_unit_area' to numeric, keeping numeric values intact
filtered_data['Price_per_unit_area'] = filtered_data['Price_per_unit_area'].apply(
    lambda x: float(x.replace(',', '')) if isinstance(x, str) else x
)
```

```python
filtered_data
```

```python
# Print unique values in the 'Price_per_unit_area' column to inspect the data
print(filtered_data['Price_per_unit_area'].unique())
```

```python
# Find the minimum and maximum values in the "Size" attribute
min_Price_per_unit_area = filtered_data['Price_per_unit_area'].min()
max_Price_per_unit_area = filtered_data['Price_per_unit_area'].max()

print("Minimum Price_per_unit_area:", min_Price_per_unit_area)
print("Maximum Price_per_unit_area:", max_Price_per_unit_area)
```

```python
# Define the price range categories and corresponding ranges
price_ranges = {
    'Budget': (100, 10000),
    'Luxury': (10000, 50000),
    'Ultra Luxury': (50000, 100000)
}

# Function to categorize price
def categorize_price(price):
    for category, (min_range, max_range) in price_ranges.items():
        if min_range <= price <= max_range:
            return category
    return 'Other'

# Group the filtered data by "City_name"
grouped_data = filtered_data.groupby('City_name')

# Initialize an empty list to store dataframes for each city
city_dataframes = []

# Iterate through each city group
for city, city_group in grouped_data:
    # Apply the categorize_price function to create the "Price_Category" column
    city_group['Price_Category'] = city_group['Price_per_unit_area'].apply(categorize_price)
    city_dataframes.append(city_group)

# Concatenate the city-specific dataframes back together
final_data = pd.concat(city_dataframes)

# Display the first few rows of the final DataFrame
print(final_data.head())
```

```python
# Find the minimum and maximum values in the "Size" attribute
min_size = final_data['Size'].min()
max_size = final_data['Size'].max()

print("Minimum Size:", min_size)
print("Maximum Size:", max_size)
```

```python
# Define the size range categories and corresponding ranges
size_ranges = {
    'Small': (100, 1000),   # Define your small size range here
    'Medium': (1000, 3000), # Define your medium size range here
    'Large': (3000, 90000)  # Define your large size range here
}

# Function to categorize size
def categorize_size(size):
    for category, (min_range, max_range) in size_ranges.items():
        if min_range <= size <= max_range:
            return category
    return 'Other'

# Group the DataFrame by "City_name" and "Price_Category"
grouped_data = final_data.groupby(['City_name', 'Price_Category'])

# Initialize an empty list to store dataframes for each group
group_dataframes = []

# Iterate through each group
for group, group_data in grouped_data:
    # Apply the categorize_size function to create the "Size_Category" column
    group_data['Size_Category'] = group_data['Size'].apply(categorize_size)
    group_dataframes.append(group_data)

# Concatenate the group-specific dataframes back together and assign it to the final_data DataFrame
final_data = pd.concat(group_dataframes)

# Display the first few rows of the final DataFrame
# print(final_data.head())
final_data
```
